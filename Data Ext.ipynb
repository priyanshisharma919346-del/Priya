```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count = 1

while True:
  URL =f"https://quotes.toscrape.com/page/{page_count}/"
  res = requests.get(URL)
  soup = BeautifulSoup(res.text,"lxml")
  quote = soup.select("div.quote")

  if not quote:
    print("invalide quote")
    break

  with open(f"sample_data/quotes{page_count}.html", "w") as f:
   f.write(res.text)
   print(f"find valide page here {page_count}")

   page_count+= 1
```
